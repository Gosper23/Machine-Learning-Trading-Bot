### Import Libraries and Dependencies

```python
# Import libraries and dependencies
import pandas as pd
import numpy as np
from pathlib import Path
%matplotlib inline

# For time stamps
from datetime import datetime

import warnings
warnings.filterwarnings('ignore')
```

### Read in CSV as Pandas DataFrame

```python
tsla_data = pd.read_csv(
    Path("../resources/technical_indicator_data.csv"),
    index_col="Date",
    parse_dates=True, 
    infer_datetime_format=True
)

tsla_data.tail()
```

### Set X-Variable List and Filter to Obtain Associated Values

```python
# Set x variable list of features
x_var_list = ['SMA_Signal', 'BB_Signal', 'EMA_Signal']

# Filter by x-variable list
tsla_data[x_var_list].tail()
```

### Shift the DataFrame Index by 1

```python
# Shift DataFrame values by 1 (this is so we can have a predictor as to whether or not it's a buy or sell)

tsla_data[x_var_list] = tsla_data[x_var_list].shift(1)
tsla_data[x_var_list].tail()
```

### Drop NAs and Replace Infs (Positive/Negative Infinity)

```python
# Drop NAs and replace positive/negative infinity values - because of the moving averages

tsla_data.dropna(subset=x_var_list, inplace=True)
tsla_data.dropna(subset=['daily_return'], inplace=True)
tsla_data = tsla_data.replace([np.inf, -np.inf], np.nan)
tsla_data.head()
```
